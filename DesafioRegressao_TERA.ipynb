## Tratamento de variáveis categóricas e numéricas

Um dos processos é diminuir a quantidade de subdivisões das variáveis categóricas, para que o nosso modelo não tenha um alto custo computacional.

```python
# Definir apenas as separações de idade até 25 anos e os valores maiores pertencerão ao mesmo grupo, maior que 25 anos.
df['TP_FAIXA_ETARIA'] = np.where(df['TP_FAIXA_ETARIA'] > 10, 11, df['TP_FAIXA_ETARIA'])
```

```python
# Classificar apenas em dois grupos, quem é 'solteiro' (que julgamos ser a maioria para um público do ENEM) 
# e o outro grupo que é composto por todas as outras opções

# Solteiro, Outro
df['TP_ESTADO_CIVIL'] = np.where(df['TP_ESTADO_CIVIL'] == 1 , 'Solteiro', 'Outro')
```

```python
# Substituir os valores numéricos de 'TP_SEXO' pelo nome correspondente
df['TP_SEXO'] = np.where(df['TP_SEXO'] == 'M' , 'Masculino', 'Feminino')
```

```python
# Diminuir agrupamento das definições de cor e raça.
# 1 e 4 = 'Branco/Amarelo'
# 2 = 'Preto'
# 3 e 5 = 'Pardo/Indigena'
# 0 e 6 = 'Não sabe/respondeu'
df['TP_COR_RACA'] = np.where(df['TP_COR_RACA'].isin([1,4]) , 'Branco/Amarelo', 
                             np.where(df['TP_COR_RACA'].isin([2]) , 'Preto', 
                                      np.where(df['TP_COR_RACA'].isin([3,5]) , 'Pardo/Indigena', 'Não sabe/respondeu')))
```

```python
# Diminuir agrupamento de nacionalidade
# 1, 2 e 4 = 'Brasileiro'
# 0 e 3 = 'Não respndeu/Estrangeiro
df['TP_NACIONALIDADE'] = np.where(df['TP_NACIONALIDADE'].isin([1,2,4]) , 'Brasileiro', 'Não respondeu/Estrangeiro')
```

```python
# Diminuir agrupamento de escolaridade
# 1 = 'Concluiu Ensino Médio'
# 0 e 3 = 'Não concluiu Ensino Médio'
df['TP_ST_CONCLUSAO'] = np.where(df['TP_ST_CONCLUSAO'] == 1 , 'Concluiu Ensino Medio', 'Não concluiu Ensino Medio')
```

```python
# Diminuir agrupamento de ano de conclusão
# 1, 2  = 'Até 2 anos'
# demais= 'Há mais de 2 anos'
df['TP_ANO_CONCLUIU'] = np.where(df['TP_ANO_CONCLUIU'].isin([1,2]) , 'Até 2 anos', 'Há mais de 2 anos')
```

```python
# Substituir os valores numéricos de 'TP_ESCOLA' pelo nome correspondente
df['TP_ESCOLA'] = np.where(df['TP_ESCOLA'].isin([2]) , 'Pública',
                            np.where(df['TP_ESCOLA'].isin([3]) , 'Privada', 'Não Respondeu'))
```

```python
# Substituir os valores numéricos de 'IN_TREINEIRO' pelo nome correspondente
df['IN_TREINEIRO'] = np.where(df['IN_TREINEIRO'].isin([1]) , 'Treineiro', 'Não treineiro')
```

```python
# Substituir os valores numéricos de 'TP_LINGUA' pelo nome correspondente
df['TP_LINGUA'] = np.where(df['TP_LINGUA'].isin([0]) , 'Inglês', 'Espanhol')
```

```python
# Diminuir agrupamento de divisão de classes por renda
# Fonte: https://cps.fgv.br/qual-faixa-de-renda-familiar-das-classes
ClasseB = ['P', 'O', 'N', 'M', 'L', 'K']
ClasseC = ['J', 'I', 'H', 'G', 'F']
ClasseD = ['E', 'D', 'C']
ClasseE = ['A', 'B']
df['Q006'] = np.where(df['Q006'].isin(ClasseE), 'Classe E', 
                          (np.where(df['Q006'].isin(ClasseD), 'Classe D',
                                   np.where(df['Q006'].isin(ClasseC), 'Classe C',
                                             np.where(df['Q006'].isin(ClasseB), 'Classe B', 'Classe A')))))
```

Agora vamos agrupar alguns tipos de perguntas em listas para aplicar mais reduções na separação e organização do dataframe para otimizar o custo computacional do modelo.


```python
#Colunas com a escolaridade dos Pais
escolaridade_cols =["Q001", "Q002"]

#Colunas com contagens de bens
questoes = ["Q007",
            "Q008",
            "Q009",
            "Q010",
            "Q011",
            "Q012",
            "Q013",
            "Q014",
            "Q015",
            "Q016",
            "Q017",
            "Q018",
            "Q019",
            "Q020",
            "Q021",
            "Q022",
            "Q023",
            "Q024",
            "Q025"] 
```

```python
# Função pra recodificar a escolaridade dos pais
Fund = ['D']
Med = ['E']
Sup = ['F', 'G']
def recode_escolaridade(df,escolaridade):
    df[escolaridade] = np.where(df[escolaridade].isin(Fund), 'Fundamental', 
                            (np.where(df[escolaridade].isin(Med), 'Médio',
                                     np.where(df[escolaridade].isin(Sup), 'Superior', 'Não sabe/respondeu'))))
```

```python
# Aplicar a função pra recodificar escolaridade dos pais
for escolaridade in escolaridade_cols:
    enem = recode_escolaridade(df,escolaridade)
```

```python
# Função pra recodificar a contagem de pertences
Um = ['B']
MaisdeUm = ['C', 'D', 'E']
def recode_counts(df,questao):
    df[questao] = np.where(df[questao].isin(Um), 'Tem 1', 
                            (np.where(df[questao].isin(MaisdeUm), 'Tem mais de 1', 'Não tem/respondeu')))
    return df
```

```python
# Aplicar a função pra recodificar as contagens
for questao in questoes:
    enem = recode_counts(df,questao)
```

## Interpretabilidade usando uma regressão linear

```python
# Vamos pegar uma amostra do nosso dado para fazer a regressão linear

sample_df = df.sample(n=500, replace=False, random_state=2023)
sample_df.shape
```

### Classe Social

Vamos fazer uma Regressão Linear simples para explicar a variável target "NU_MEDIA" pela variável "Q006".

Legenda:
Variável target "NU_MEDIA": notá do aluno no ENEM.
Variável explicativa "Q006": Qual a classe social do aluno

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Ajuste do Modelo simples
modelo_rl1 = smf.ols(formula = 'NU_MEDIA ~ Q006', data = sample_df)
modelo_rl1_fit = modelo_rl1.fit()
modelo_rl1_fit.summary()
```

R² de 0,275 indica que cerca de 27,5% da variação observada na nota do aluno no enem (target) pode ser explicada pela classe social dele.

### Cor / Raça

```python
# Ajuste do Modelo simples
modelo_rl2 = smf.ols(formula = 'NU_MEDIA ~ TP_COR_RACA', data = sample_df)
modelo_rl2_fit = modelo_rl2.fit()
modelo_rl2_fit.summary()
```

R² de 0,061 indica que cerca de 6,1% da variação observada na nota do aluno no enem (target) pode ser explicada pela cor ou raça dele.

### Escola Pública ou Privada

```python
# Ajuste do Modelo simples
modelo_rl3 = smf.ols(formula = 'NU_MEDIA ~ TP_ESCOLA', data = sample_df)
modelo_rl3_fit = modelo_rl3.fit()
modelo_rl3_fit.summary()
```

R² de 0,082 indica que cerca de 8,2% da variação observada na nota do aluno no enem (target) pode ser explicada pelo tipo de escola, públcia ou privada, ele estudou.

### Concluiu ou não o Ensino Médio

```python
# Ajuste do Modelo simples
modelo_rl4 = smf.ols(formula = 'NU_MEDIA ~ TP_ST_CONCLUSAO', data = sample_df)
modelo_rl4_fit = modelo_rl4.fit()
modelo_rl4_fit.summary()
```

R² de 0,030 indica que cerca de 3% da variação observada na nota do aluno no enem (target) pode ser explicada pelo fato dele ter concluido, ou não, o ensino médio.

### Grau de Escolaridade do Pai

```python
# Ajuste do Modelo simples
modelo_rl5 = smf.ols(formula = 'NU_MEDIA ~ Q001', data = sample_df)
modelo_rl5_fit = modelo_rl5.fit()
modelo_rl5_fit.summary()
```

R² de 0,153 indica que cerca de 15,3% da variação observada na nota do aluno no enem (target) pode ser explicada pelo fato do seu pai ter concluido, ou não, o ensino médio.

### Grau de Escolaridade da Mãe

```python
# Ajuste do Modelo simples
modelo_rl6 = smf.ols(formula = 'NU_MEDIA ~ Q002', data = sample_df)
modelo_rl6_fit = modelo_rl6.fit()
modelo_rl6_fit.summary()
```

R² de 0,129 indica que cerca de 12,9% da variação observada na nota do aluno no enem (target) pode ser explicada pelo fato da sua mãe ter concluido, ou não, o ensino médio.

### Conclusão

#### Maiores correlações

Os três maiores valores de R² foram: classe social, escolaridade do pai e escolaridade da mãe, respectivamente. Classe social teve um desempenho consideravelmente maior que os dois seguintes, obtendo um R² de 0,275.

De toda forma, podemos entender que todas as variáveis explicativas utilizadas estão relacionadas a diferenças sociais, onde nem todos podem ter acesso a um ensino de qualidade, o que automaticamente afeta a nota deste aluno na prova do ENEM.

#### O valor do R²

Um R² de 0,275 pode ser considerado moderado a baixo, dependendo da área em que o modelo está sendo aplicado. No caso de modelagem de comportamentos humanos, pode ser mais difícil obter altos valores de R², uma vez que os seres humanos são altamente complexos e influenciados por uma ampla gama de fatores que não podem ser facilmente quantificados ou medidos.

No contexto específico de modelagem do desempenho de alunos em provas do ENEM, um R² de 0,275 pode indicar que a variável categórica explicativa usada em seu modelo (como a classe social do aluno, por exemplo) tem uma influência moderada na variação do desempenho dos alunos. Isso significa que outros fatores além desta variável categórica explicativa (como escolaridade dos pais, por exemplo) podem ter um efeito significativo no desempenho dos alunos, e que esses fatores também devem ser levados em consideração ao tentar prever ou explicar o desempenho dos alunos.

É importante lembrar que a qualidade do modelo depende não apenas do valor de R², mas também de outros fatores, como a significância estatística das variáveis explicativas, a precisão das previsões do modelo e a adequação do modelo às suposições subjacentes. Portanto, é importante avaliar todos esses aspectos ao interpretar os resultados do modelo.

Neste caso, o objetivo era identificar uma variável explicativa que tivesse uma maior influência na nota do aluno no ENEM (target), por isso, o R² foi avaliado.

```python
from scipy import stats
from scipy.stats import shapiro

# Gráfico QQ - Normalidade
modelo_residuo = modelo_rl1_fit.resid
modelo_y_ajustado = modelo_rl1_fit.fittedvalues

sm.qqplot(modelo_residuo, line="45", fit=True,dist=stats.norm)
plt.show()

# Teste Shapiro-Wilk para normalidade dos residuos
stat, p = shapiro(modelo_residuo)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# Interpretação
alpha = 0.05
if p > alpha:print('A amostra parece Normal (Não rejeitamos H0)')
else:print('A amostra não parece Normal (rejeitamos H0)')
```

#### Shapiro-Wilk

O teste de normalidade de Shapiro-Wilk é utilizado para testar a hipótese nula de que uma amostra de dados segue uma distribuição normal. No caso da regressão linear, é comum testar a normalidade dos resíduos do modelo, ou seja, dos erros de previsão. Se os resíduos não seguem uma distribuição normal, isso pode indicar que o modelo não está capturando toda a variação dos dados e que as previsões podem estar enviesadas.

O p-valor do teste de Shapiro-Wilk indica a probabilidade de obter os dados observados, ou dados ainda mais extremos, sob a hipótese nula de normalidade. Se o p-valor for maior que o nível de significância escolhido (geralmente 0,05), não há evidência estatística suficiente para rejeitar a hipótese nula de normalidade. Isso significa que a amostra parece normal e que os resíduos do modelo seguem uma distribuição normal. Neste caso, o resultado "A amostra parece Normal (Não rejeitamos H0)" indica que os resíduos do modelo de regressão linear seguem uma distribuição normal.

Isso sugere que o modelo está capturando adequadamente a variação dos dados e que as previsões não estão enviesadas devido a uma violação da suposição de normalidade dos resíduos. No entanto, é importante lembrar que a normalidade dos resíduos não é a única suposição da regressão linear e que outras suposições, como a homogeneidade das variâncias e a linearidade da relação entre as variáveis, também devem ser avaliadas.

#### Custo Computacional

O uso de One-Hot Encoding pode aumentar significativamente o custo computacional e a dimensão do conjunto de dados. Isso ocorre porque, para cada categoria de uma variável categórica, o One-Hot Encoding cria uma nova coluna binária. Se a variável categórica tiver muitas categorias, isso pode levar a um grande número de colunas e a um conjunto de dados muito maior.

Além disso, o uso de One-Hot Encoding pode levar a problemas de esparsidade e multicolinearidade. Se uma variável categórica tiver muitas categorias, muitas das novas colunas binárias geradas pelo One-Hot Encoding podem ter apenas um valor "1" em todo o conjunto de dados. Isso significa que a maioria das colunas terá valores "0" para a maioria das amostras, levando a um conjunto de dados esparsos. Além disso, as novas colunas geradas pelo One-Hot Encoding podem ser altamente correlacionadas entre si, o que pode levar a problemas de multicolinearidade em modelos de regressão.

Por isso, é importante avaliar o impacto do One-Hot Encoding no tamanho do conjunto de dados e na qualidade dos modelos antes de decidir usá-lo. Em alguns casos, outras técnicas de codificação de variáveis categóricas, como a codificação ordinal, podem ser mais apropriadas e eficientes.

#### Sobre as métricas de avaliação

O MSE (Erro Quadrático Médio) é uma medida de desempenho que indica o quão distantes os valores previstos pelo modelo estão dos valores reais. Quanto menor o valor do MSE, melhor é o desempenho do modelo, porém, ele não está na mesma unidade de medida que a variável target, o que pode dificultar sua interpretação em relação a sua grandeza.

O RMSE (Raiz do Erro Quadrático Médio) é uma medida de desempenho semelhante ao MSE, mas que expressa o erro médio em unidades de medida da variável target. Quanto menor o valor do RMSE, melhor é o desempenho do modelo.

O MAE (Erro Absoluto Médio) é outra medida de desempenho que indica o quão distantes os valores previstos pelo modelo estão dos valores reais, mas sem levar em consideração o sinal da diferença. Ou seja, o MAE indica o quão longe, em média, as previsões estão dos valores reais, mas não se importa se o modelo superestimou ou subestimou a nota do aluno. Quanto menor o valor do MAE, melhor é o desempenho do modelo.
